```python
import pandas as pd
```

```python
df=pd.read_csv('car data.csv')
```

```python
df.isnull().sum()
```

## Performing Feature Engineering 

```python
#1.Removing features with no use like car_name
df.columns
```

```python
new_df=df[[ 'Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
new_df.head()
```

```python
#2.Creating new feature "Year"

new_df['Current_year']=2020
new_df['Old']=new_df['Current_year']-new_df['Year']
```

```python
new_df.head()
```

```python
#3.Dropping some more unneccessary coloums
new_df.drop(['Year'],axis=1,inplace=True)
new_df.drop(['Current_year'],axis=1,inplace=True)

new_df.head()
```

```python
#Identifing categorical variables 
print(df['Fuel_Type'].unique(),'\n', df['Seller_Type'].unique(),'\n', df['Transmission'].unique(),'\n', df['Owner'].unique())
```

```python
#4.Converting Categorical fetures into Numerical variable using one-hot encoding 

new_df=pd.get_dummies(new_df,drop_first=True)
new_df.head()
```

## Feature Selection ##

```python
corr=new_df.corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(figsize=(15, 15))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
heatmap = sns.heatmap(corr,cmap=cmap, center=0.0,vmax = 1, linewidths=1, ax=ax,annot=True)
plt.show()
```

## Model Creation 

```python
new_df.head()
```

```python
#x=independent features
#y=dependent features

x=new_df.iloc[:,1:]
y=new_df.iloc[:,:1]
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
print(new_df.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

## Creating Pickle File

```python
import pickle
file = open('lr_model.pkl','wb')
pickle.dump(model,file)
```
